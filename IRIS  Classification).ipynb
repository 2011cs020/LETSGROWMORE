# Importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

# Data Loading

```python
#Here we give name to each columns 
names=['sepal length in cm',
   'sepal width in cm',
   'petal length in cm',
   'petal width in cm', 'class']
#we read data seet as set headers to our data
data=pd.read_csv('iris.data',names=names)
```

# Feature selection

```python
#selecting dependent and independent variables
X=data[['sepal length in cm','sepal width in cm', 'petal length in cm','petal width in cm']]
y=data['class']
```

# Model Selection

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
# Create a Random Forest classifier
target = RandomForestClassifier(n_estimators=100, random_state=42)



# Train the model
target.fit(X_train, y_train)
y_pred = target.predict(X_test)
```

# Model Evaluation

```python

print(accuracy_score(y_test,y_pred))
```

# from sklearn.metrics import classification_report

```python
report = classification_report(y_test, y_pred)
print(report)
```
